```python
# Gerekli kütüphanelerin yüklenmesi
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc,mean_squared_error,accuracy_score
```

```python
df=pd.read_csv('veriseti.csv')
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

-bir DataFrame'in her sütundaki eksik değerlerin sayısını hesaplar.

```python
df.info() #Veri setindeki değerler hakkında bilgi almak için kullanıyoruz.
```

```python
df.eq(0).sum() #Her bir özellikte kaç tane 0 var onu öğrenmek için kullanıyoruz.
```

```python
df[['Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']] = df[['Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']].replace(0,np.NaN)
```

```python
# Eksik değerlerin doldurulması için kullanıyoruz.
df.fillna(df.mean(), inplace=True)
```

```python
# Normalizasyon
normalized_df = (df - df.min()) / (df.max() - df.min())
print(normalized_df.head(6))
```

-Bu kod, bir DataFrame'deki değerlerin normalizasyonunu gerçekleştirir. Normalizasyon, veri kümesindeki değerleri bir aralığa (genellikle [0, 1] veya [-1, 1] aralığına) dönüştürme işlemidir. Bu, verilerin birbirleriyle karşılaştırılabilir hale gelmesini sağlar ve bazı makine öğrenimi algoritmaları için önemlidir.

```python
# Bağımsız değişkenlerin ve hedef değişkenin belirlenmesi için kullanıyoruz.
X = df.drop("Outcome", axis=1)
y = df["Outcome"]
```

-Bu kod, tipik bir makine öğrenimi modeli için girdi (X) ve çıktı (y) değişkenlerini ayırmak için kullanılır. Genellikle, makine öğrenimi modelleri bir hedef değişkeni tahmin etmek için bağımsız değişkenleri kullanır. 

```python
sc=StandardScaler()
x1=sc.fit_transform(X)
pca=PCA(n_components=2)
PrincipialComponent=pca.fit_transform(x1)
principialDF=pd.DataFrame(data=PrincipialComponent,columns=["p1","p2"])
principialDF.head()
```


-Bu kod, bir veri setindeki özelliklerin (bağımsız değişkenlerin) boyutunu azaltmak için bir dizi dönüşüm işlemi uygular. Genellikle bu tür işlemler, veri görselleştirmesi veya daha az boyutlu veri kümesiyle makine öğrenimi modelleri için çalışma yapmak amacıyla gerçekleştirilir.

```python
# LDA uygulama
lda = LDA(n_components=1)
X_lda = lda.fit_transform(X, y)
```

-Bu kod, Linear Discriminant Analysis (LDA) yöntemini kullanarak veri kümesinin boyutunu azaltmayı amaçlar. LDA, özelliklerin boyutunu azaltmak ve sınıflar arasındaki farkı maksimize etmek için yaygın olarak kullanılan bir tekniktir

```python
df_lda = pd.DataFrame(X_lda, columns=['LD1'])
print(df_lda.head())
```

```python
# Veri setinin eğitim ve test setlerine ayrılması için kullanıyoruz.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

-Bu kod, veri setini eğitim ve test setlerine ayırmak için train_test_split fonksiyonunu kullanır. Bu işlem, genellikle makine öğrenimi modellerini eğitmek ve değerlendirmek için kullanılan bir yöntemdir.

```python
# Çoklu Doğrusal Regresyon analizi için kullanıyoruz.
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)
y_pred_linear_reg = linear_reg.predict(X_test)
linear_reg_coefficients = linear_reg.coef_
```

-Bu kod, çoklu doğrusal regresyon analizi için bir model oluşturur ve bu modeli eğitir. Daha sonra, bu eğitilmiş modeli kullanarak test setinde tahminler yapar ve regresyon katsayılarını elde eder. 

```python
# Multinominal Lojistik Regresyon analizi için kullanıyoruz.
logistic_reg = LogisticRegression(multi_class='multinomial', solver='lbfgs')
logistic_reg.fit(X_train, y_train)
y_pred_logistic_reg = logistic_reg.predict(X_test)
```


-Bu kod, multinominal lojistik regresyon analizi için bir model oluşturur ve bu modeli eğitir. 

```python
# Karar Ağacı sınıflandırma algoritması için kullanıyoruz.
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, y_train)
y_pred_decision_tree = decision_tree.predict(X_test)
```

-Bu kod, karar ağacı sınıflandırma algoritması için bir model oluşturur ve bu modeli eğitir. 

```python
# Naive Bayes sınıflandırıcısı için kullanıyoruz.
naive_bayes = GaussianNB()
naive_bayes.fit(X_train, y_train)
y_pred_naive_bayes = naive_bayes.predict(X_test)
```


-Bu kod, Naive Bayes sınıflandırıcı modeli için veri setini eğitir ve bu modeli kullanarak test setinde tahminler yapar.

```python
# Performans metriklerinin hesaplanması için kullanıyoruz.
linear_reg_metrics = {
    "MSE": mean_squared_error(y_test, y_pred_linear_reg)
}
```


-Bu kod, doğrusal regresyon modelinin performansını değerlendirmek için bir metrik olan ortalama karesel hata (MSE) değerini hesaplar.

```python
logistic_reg_metrics = {
    "Accuracy": accuracy_score(y_test, y_pred_logistic_reg),
    "Classification Report": classification_report(y_test, y_pred_logistic_reg),
    "Confusion Matrix": confusion_matrix(y_test, y_pred_logistic_reg)
}
```


-Bu kod, lojistik regresyon modelinin performansını değerlendirmek için bir dizi metrik hesaplar. 

```python
decision_tree_metrics = {
    "Accuracy": accuracy_score(y_test, y_pred_decision_tree),
    "Classification Report": classification_report(y_test, y_pred_decision_tree),
    "Confusion Matrix": confusion_matrix(y_test, y_pred_decision_tree)
}
```


-Bu kod, karar ağacı sınıflandırıcının performansını değerlendirmek için bir dizi metrik hesaplar. 

```python
naive_bayes_metrics = {
    "Accuracy": accuracy_score(y_test, y_pred_naive_bayes),
    "Classification Report": classification_report(y_test, y_pred_naive_bayes),
    "Confusion Matrix": confusion_matrix(y_test, y_pred_naive_bayes)
}
```

-Bu kod, Naive Bayes sınıflandırıcının performansını değerlendirmek için bir dizi metrik hesaplar. 

```python
# Sonuçların raporlanması için kullanıyoruz.
print("Çoklu Doğrusal Regresyon Analizi Sonuçları:")
print("Katsayılar:", linear_reg_coefficients)
print("Mean Squared Error (MSE):", linear_reg_metrics["MSE"])
print()
```

Bu çıktı, doğrusal regresyon analizinin sonuçlarını içerir. Şimdi sayısal çıktıların yorumunu yapalım:

**Katsayılar (Coefficients)**: Her bir katsayı, bağımsız değişkenlerin (predictors) etkisini temsil eder. Örneğin, "9.58400584e-03" olan bir katsayı, ilgili bağımsız değişkenin bir birim artışının hedef değişkeni 0.00958 birim artıracağını önerir. Pozitif katsayılar, bağımsız değişkenin artmasıyla hedef değişkenin artacağını, negatif katsayılar ise bağımsız değişkenin artmasıyla hedef değişkenin azalacağını gösterir.

**Mean Squared Error (MSE) (Ortalama Karesel Hata)**: MSE, modelin gerçek ve tahmin edilen değerleri arasındaki ortalama karesel farkı gösterir. "0.174317237562553" olan bu değer, modelin tahminlerinin gerçek değerlerden ortalama olarak ne kadar uzak olduğunu belirtir. Düşük MSE değerleri, daha iyi bir model performansını işaret eder.

```python
print("Multinominal Lojistik Regresyon Analizi Sonuçları:")
print("Accuracy:", logistic_reg_metrics["Accuracy"])
print("Classification Report:\n", logistic_reg_metrics["Classification Report"])
print("Confusion Matrix:\n", logistic_reg_metrics["Confusion Matrix"])
print()
```

multinominal lojistik regresyon analizinin sonuçlarını içerir. Şimdi sayısal çıktıların yorumunu yapalım:

**Accuracy (Doğruluk)**: Bu, modelin doğru sınıflandırma oranını ifade eder. "0.7402597402597403" olan bu değer, modelin doğru sınıflandırma oranının %74 olduğunu gösterir. Yani, tüm tahminlerin yaklaşık olarak %74'ünün doğru olduğu anlamına gelir.

**Classification Report (Sınıflandırma Raporu)**: Bu bölüm, her sınıf için precision (kesinlik), recall (duyarlılık) ve f1-score (F1 skoru) gibi metrikleri sunar. Bunlar, modelin sınıflandırma performansını daha ayrıntılı olarak değerlendirmek için kullanılır. Ayrıca, her sınıf için destek (support) değerleri de verilir. Örneğin, (0, 0) hücresi, gerçek sınıfın 0 ve tahmin edilen sınıfın da 0 olduğu durumları ifade eder. Bu matriste köşegen dışındaki değerler yanlış sınıflandırmaları gösterir.
